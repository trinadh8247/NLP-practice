```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
data = pd.read_csv(path + '/spam.csv', encoding='latin-1')
# print(data.head())
data = data[['v1', 'v2']]
data.columns = ['label', 'message']
```

```python
import nltk
nltk.download('punkt')
import re
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
corpus=[]
lemmatizer = WordNetLemmatizer()
for i in range(0,len(data)):
  review = re.sub('[^a-zA-Z]', ' ', data['message'][i])
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review]
  review = ' '.join(review)
  corpus.append(review)
```

```python
# Filter out empty strings from the corpus
filtered_corpus = [[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,data['message']) if i<1]
filtered_corpus
```

```python
from nltk import sent_tokenize
nltk.download('punkt_tab')
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
  sent_tokenized = sent_tokenize(sent)
  for sent in sent_tokenized:
    words.append(simple_preprocess(sent))
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
  # Filter out words not in the vocabulary
  valid_words = [word for word in doc if word in model.wv.index_to_key]
  if not valid_words:
    # Return a zero vector of size 100 if no valid words are found
    return np.zeros(model.wv.vector_size)
  return np.mean([model.wv[word] for word in valid_words], axis=0)
```
